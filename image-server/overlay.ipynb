```python
import cv2
import numpy as np
import matplotlib.pylab as plt

# Load the image
image = cv2.imread('./preprocessed/normal/0a0d42ce-6e76-4ee0-afa8-65d93b94771e.png')
plt.imshow(image)

# Rectangle coordinates
top_left = (150, 250)  # Change these values as per your requirement
bottom_right = (300, 400)  # Change these values as per your requirement
color = (0, 255, 0)  # Rectangle color (Green in this case)
alpha = 0.4  # Transparency factor

# Make a copy of the image
overlay = image.copy()

# Draw a filled rectangle on the copy
cv2.rectangle(overlay, top_left, bottom_right, color, -1)

# Blend the overlay with the original image
cv2.addWeighted(overlay, alpha, image, 1 - alpha, 0, image)

plt.imshow(image)
```
